```python
import pygame
# görüntü yüzeyi oluşturma
pygame.init()

en = 800
boy = 600

goruntu_yuzeyi = pygame.display.set_mode((en, boy))
# görüntü kopyalama
karakter = pygame.image.load("monster.png")
karakter_koordinat = karakter.get_rect() # dikdörtgen koordinatlarına sahip bir obje oluşturuldu
karakter_koordinat.topleft = (100,100) # 0,0 noktasında karakter oluşturuldu

deger = 50

durum = True
while durum:
    
    for etkinlik in pygame.event.get():
        if etkinlik.type == pygame.QUIT:
            durum = False
            
        if etkinlik.type == pygame.KEYDOWN: #klavyedeki tuşlara basarsan
            if etkinlik.key == pygame.K_LEFT: #sol tuşuna basarsan
                karakter_koordinat.x -= deger #x koordinatını deger kadar azalt
            if etkinlik.key == pygame.K_RIGHT: #sag tuşuna basarsan
                karakter_koordinat.x += deger
            if etkinlik.key == pygame.K_UP: #yukarı tuşuna basarsan
                karakter_koordinat.y -= deger
            if etkinlik.key == pygame.K_DOWN: #asagi tuşuna basarsan
                karakter_koordinat.y += deger
            
                
            
    # arka plan temizleme
    goruntu_yuzeyi.fill((255,0,255))        
    # görüntü kopyalama       
    goruntu_yuzeyi.blit(karakter,karakter_koordinat) # goruntu yüzeyine karakteri ekle
    
    
    pygame.display.update()
    

pygame.quit()
```

```python
import pygame
# görüntü yüzeyi oluşturma
pygame.init()

en = 800
boy = 600

goruntu_yuzeyi = pygame.display.set_mode((en, boy))
# görüntü kopyalama
karakter = pygame.image.load("monster.png")
karakter_koordinat = karakter.get_rect() # dikdörtgen koordinatlarına sahip bir obje oluşturuldu
karakter_koordinat.topleft = (100,100) # 0,0 noktasında karakter oluşturuldu

FPS = 10 # frame per second=yenileme hızı 
fps_saat = pygame.time.Clock()

deger = 50

durum = True
while durum:
    
    for etkinlik in pygame.event.get():
        #print(etkinlik)
        if etkinlik.type == pygame.QUIT:
            durum = False
        
    tus = pygame.key.get_pressed() # basılan tuşlara eriştik
    if tus[pygame.K_LEFT] and karakter_koordinat.left > 0:
        karakter_koordinat.x -= deger
    elif tus[pygame.K_RIGHT] and karakter_koordinat.right < en:
        karakter_koordinat.x += deger
    elif tus[pygame.K_UP] and karakter_koordinat.top > 0:
        karakter_koordinat.y -= deger
    elif tus[pygame.K_DOWN] and karakter_koordinat.bottom < boy:
        karakter_koordinat.y += deger
            
            
    # arka plan temizleme
    goruntu_yuzeyi.fill((255,0,255))
        
    # görüntü ekleme       
    goruntu_yuzeyi.blit(karakter,karakter_koordinat) # goruntu yüzeyine karakteri ekle

    # güncelleme
    pygame.display.update()
    # FPS
    fps_saat.tick(FPS)
    

pygame.quit()
```

```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
width, height = 1000, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Space Dodger")

# Colors
black = (0, 0, 0)
white = (255, 255, 255)
red = (255, 0, 0)
green = (255, 0, 0)

# Player (spaceship) settings
player_size = 50
player_pos = [width // 2, height - 2 * player_size]
player_speed = 10

# Enemy (asteroid) settings
enemy_size = 50
enemy_pos = [random.randint(0, width - enemy_size), 0]
enemy_speed = 5

# Game variables
score = 0
lives = 3
level = 1

# Clock for controlling frame rate
clock = pygame.time.Clock()

font = pygame.font.SysFont("monospace", 35)

# Function to drop enemies
def drop_enemies(enemy_list):
    delay = random.random()
    if len(enemy_list) < 10 and delay < 0.1:
        x_pos = random.randint(0, width - enemy_size)
        y_pos = 0
        enemy_list.append([x_pos, y_pos])

# Function to update enemy positions
def update_enemy_positions(enemy_list, score):
    for idx, enemy_pos in enumerate(enemy_list):
        if enemy_pos[1] >= 0 and enemy_pos[1] < height:
            enemy_pos[1] += enemy_speed
        else:
            enemy_list.pop(idx)
            score += 1
    return score

# Function to detect collisions
def collision_check(enemy_list, player_pos):
    for enemy_pos in enemy_list:
        if detect_collision(enemy_pos, player_pos):
            return True
    return False

# Function to detect collision
def detect_collision(enemy_pos, player_pos):
    p_x, p_y = player_pos
    e_x, e_y = enemy_pos

    if (e_x >= p_x and e_x < (p_x + player_size)) or (p_x >= e_x and p_x < (e_x + enemy_size)):
        if (e_y >= p_y and e_y < (p_y + player_size)) or (p_y >= e_y and p_y < (e_y + enemy_size)):
            return True
    return False

# Main game loop
game_over = False
enemy_list = []

while not game_over:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            game_over = True

    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT] and player_pos[0] > 0:
        player_pos[0] -= player_speed
    if keys[pygame.K_RIGHT] and player_pos[0] < width - player_size:
        player_pos[0] += player_speed

    screen.fill(black)

    drop_enemies(enemy_list)
    score = update_enemy_positions(enemy_list, score)
    text = "Score: {}  Lives: {}  Level: {}".format(score, lives, level)
    label = font.render(text, 1, white)
    screen.blit(label, (width - 700, height - 40))

    if collision_check(enemy_list, player_pos):
        lives -= 1
        if lives == 0:
            game_over = True
        else:
            player_pos = [width // 2, height - 2 * player_size]
            enemy_speed = 5
            enemy_list = []

    if score > 0 and score % 10 == 0:
        level += 1
        enemy_speed += 1
        score += 1

    for enemy_pos in enemy_list:
        pygame.draw.rect(screen, green, (enemy_pos[0], enemy_pos[1], enemy_size, enemy_size))

    pygame.draw.rect(screen, red, (player_pos[0], player_pos[1], player_size, player_size))
    clock.tick(30)
    pygame.display.update()

pygame.quit()
```

```python
import pygame  # Pygame modülünü içe aktar
import random  # Rastgele modülünü içe aktar

pygame.init()  # Pygame'i başlat

genislik = 1200  # Oyun ekranının genişliği
yukseklik = 700  # Oyun ekranının yüksekliği
goruntu_yuzeyi = pygame.display.set_mode((genislik, yukseklik))  # Oyun ekranını ayarla

FPS = 30  # Saniye başına düşen kare sayısı
saat = pygame.time.Clock()  # Saati ayarla

class Oyun():  # Oyun sınıfı
    def __init__(self, oyuncu, canavar_grubu): 
        self.skor = 0  # Başlangıç skoru
        self.round_numarasi = 0

        self.round_suresi = 0
        self.cerceve_sayisi = 0

        self.oyuncu = oyuncu
        self.canavar_grubu = canavar_grubu

        self.ileri_seviye_muzik = pygame.mixer.Sound("arka_ses.wav")
        self.font = pygame.font.Font("AttackGraffiti.ttf", 30)

        mavi_canavar = pygame.image.load("mavi_canavar.webp")
        mavi_canavar = pygame.transform.scale(mavi_canavar, (64,64))
        yesil_canavar = pygame.image.load("yesil_canavar.webp")
        yesil_canavar = pygame.transform.scale(yesil_canavar, (64,64))
        kirmizi_canavar = pygame.image.load("monster.png")
        kirmizi_canavar = pygame.transform.scale(kirmizi_canavar, (64,64))
        turuncu_canavar = pygame.image.load("turuncu_canavar.webp")
        turuncu_canavar = pygame.transform.scale(turuncu_canavar, (64,64))

        self.canavarlar = [mavi_canavar, yesil_canavar, kirmizi_canavar, turuncu_canavar]
        self.canavar_tipi = random.randint(0,3)
        self.canavar_resmi = self.canavarlar[self.canavar_tipi]

        self.canavar_koordinati = self.canavar_resmi.get_rect()
        self.canavar_koordinati.centerx = genislik//2
        self.canavar_koordinati.top = 40

    def update(self):
        self.cerceve_sayisi += 1
        if self.cerceve_sayisi == FPS:
            self.round_suresi += 1
            self.cerceve_sayisi = 0

        self.carpisma_tespiti()
  
        
        
    def cizmek(self):
        
        beyaz = (255, 255, 255)
        mavi = (0, 0, 255)
        yesil = (0, 255, 0)
        turuncu = (255, 128, 0)
        kirmizi = (255, 0, 0)
        
        renkler = [mavi, kirmizi, yesil, turuncu]
        
        yakalama_metni = self.font.render("YAKALANACAK CANAVAR", True, beyaz)
        yakalama_koordinati = yakalama_metni.get_rect()
        yakalama_koordinati.centerx = genislik//2
        yakalama_koordinati.top = 10
        
        skor_metni = self.font.render(f"SKOR: {self.skor}", True, beyaz)
        skor_koordinati = skor_metni.get_rect()
        skor_koordinati.topleft = (10, 10)
      
        can_metni = self.font.render(f"CAN: {self.oyuncu.can}", True, beyaz)
        can_koordinati = can_metni.get_rect()
        can_koordinati.topleft = (10, 50)
       
        round_metni = self.font.render(f"ROUND: {self.round_numarasi}", True, beyaz)
        round_koordinati = round_metni.get_rect()
        round_koordinati.topleft = (10, 70)
        
        zaman_metni = self.font.render(f"ROUND SURESI: {self.round_suresi}", True, beyaz)
        zaman_koordinati = zaman_metni.get_rect()
        zaman_koordinati.topright = (genislik-10, 10)
       
        joker_metni = self.font.render(f"JOKERLER: {self.oyuncu.jokerler}", True, beyaz)
        joker_koordinati = joker_metni.get_rect()
        joker_koordinati.topright = (genislik-10, 50)
       
        goruntu_yuzeyi.blit(self.canavar_resmi, self.canavar_koordinati)
        
        goruntu_yuzeyi.blit(yakalama_metni, yakalama_koordinati)
        goruntu_yuzeyi.blit(skor_metni, skor_koordinati)
        goruntu_yuzeyi.blit(can_metni, can_koordinati)
        goruntu_yuzeyi.blit(round_metni, round_koordinati)
        goruntu_yuzeyi.blit(zaman_metni, zaman_koordinati)
        goruntu_yuzeyi.blit(joker_metni, joker_koordinati)
        
        pygame.draw.rect(goruntu_yuzeyi, renkler[self.canavar_tipi], 
                         (genislik//2-32, 40, 64, 64),2)
        pygame.draw.rect(goruntu_yuzeyi, renkler[self.canavar_tipi],
                         (0, 100, genislik, yukseklik-200),4)
    
    def carpisma_tespiti(self):
        carpisan_canavar = pygame.sprite.spritecollideany(self.canavar_grubu, self.oyuncu)
        
        if carpisan_canavar.type == self.canavar_tipi:
            self.skor += 100 * self.round_numarasi
            carpisan_canavar.remove(self.canavar_grubu)
            
            if self.canavar_grubu:
                self.oyuncu.yakalama_sesi()
                self.yeni_hedef_secenegi()
                
            else:
                self.oyuncu.yenileme()
                self.yeni_round_baslama()
                
        else:
            self.oyuncu.yanma_sesi.play()
            self.oyuncu.can -= 1
            
            if self.oyuncu.can <= 0:
                self.oyun_durdu(f"OYUN SKORU: {self.skor}, LÜTFEN ENTER TUŞUNA BASIN")
                self.oyun_yenileme()
            self.oyuncu.yenileme()
            
    def yeni_round_baslama(self):
        #Round sayısı arttıkça daha yüksek puan kazandığımız ve round süresini kısa tutarak daha yüksek puan kazandığımız formül
        self.skor += int(10000 * self.round_numarasi / (1+self.round_suresi)) 
        
        self.round_numarasi += 1
        self.cerceve_sayisi = 0
        self.round_suresi = 0
        self.oyuncu.jokerler += 1
        
        for canavar in self.canavar_grubu:
            self.canavar_grubu.remove(canavar)
            
        for i in range(self.round_numarasi):
            self.canavar_grubu.add(
            Canavar(random.randint(0, genislik-64), random.randint(0, yukseklik-164),
                    self.canavar_resmi[0], 0))
            
            self.canavar_grubu.add(
            Canavar(random.randint(0, genislik-64), random.randint(0, yukseklik-164),
                    self.canavar_resmi[1], 1))
            
            self.canavar_grubu.add(
            Canavar(random.randint(0, genislik-64), random.randint(0, yukseklik-164),
                    self.canavar_resmi[2], 2))
            
            self.canavar_grubu.add(
            Canavar(random.randint(0, genislik-64), random.randint(0, yukseklik-164),
                    self.canavar_resmi[3], 3))
            
            self.yeni_hedef_secenegi()
            
            self.ileri_seviye_muzik.play()
            
    def yeni_hedef_secenegi():
        hedef_canavar = random.choice(self.canavar_grubu.sprites())
        self.canavar_tipi = hedef_canavar.type
        self.canavar_resmi = hedef_canavar.image
        
    def oyun_durdu(self, ana_metin, baslik_metni):
        global durum
        
        beyaz = (255, 255, 255)
        siyah = (0, 0, 0)
        
        ana_metin = self.font.render(ana_metin, True, beyaz)
        ana_metin_koordinat = ana_metin.get_rect()
        ana_metin_koordinat.center = (genislik//2, yukseklik//2)
        
        baslik_metni = self.font.render(baslik_metni, True, siyah)
        baslik_metni_koordinat = baslik_metni.get_rect()
        baslik_metni_koordinat.center = (genislik//2, yukseklik//2-64)
        
        goruntu_yuzeyi.fill("black")
        
        goruntu_yuzeyi.blit(baslik_metni, baslik_metni_koordinat)
        goruntu_yuzeyi.blit(ana_metin, ana_metin_koordinat)
        pygame.display.update()
        
        durdu=True
        while durdu:
            for etkinlik in pygame.event.get():
                if etkinlik.type == pygame.KEYDOWN:
                    if etkinlik.key == pygame.K_RETURN: # Enter tuşuna basıldığında durma durumunu sonlandırır.
                        durdu = False
                if etkinlik.type == pygame.QUIT:
                    durdu = False
                    durum = False
                    

    def oyun_yenileme(self):
        self.skor = 0
        self.round_numarasi = 0
        
        self.oyuncu.can = 5
        self.oyuncu.jokerler = 3
        self.oyuncu.yenileme()
        
        self.yeni_round_baslama()
        
class Oyuncu(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.image.load("protector.png")
        self.rect = self.image.get_rect()
        self.rect.center = (genislik//2, yukseklik)
        
        self.can = 5
        self.jokerler = 3
        self.hiz = 8
        
        self.yakalama_sesi = pygame.mixer.Sound("mouse-click-sound.mp3")
        self.joker_sesi = pygame.mixer.Sound("eb_altn_ses.wav")
        self.yanma_sesi = pygame.mixer.Sound("eb_yanma_ses.wav")
        
    def update(self):
        tuslar = pygame.key.get_pressed()
        
        if tuslar[pygame.K_LEFT] and self.rect.left > 0:
            self.rect.x -= self.hiz
        if tuslar[pygame.K_RIGHT] and self.rect.right < genislik:
            self.rect.x += self.hiz
        if tuslar[pygame.K_UP] and self.rect.top > 100:
            self.rect.y -= self.hiz
        if tuslar[pygame.K_DOWN] and self.rect.bottom < yukseklik-100:
            self.rect.y += self.hiz
       
    def joker(self):
        if self.jokerler > 0:
            self.jokerler -= 1
            self.joker_sesi.play()
            self.rect.bottom = yukseklik
    
        def yenileme(self):
            self.rect.center = (genislik//2, yukseklik)
        
        
#class Canavar(pygame.sprite.Sprite):
    #def __init__(self, x, y, image, canavar_tipi)
            

oyun = Oyun()
durum = True 
while durum:  # Ana oyun döngüsü
    for etkinlik in pygame.event.get():
        if etkinlik.type == pygame.QUIT:  # Eğer çıkış istenirse
            durum = False

    goruntu_yuzeyi.fill("orange") 
    
    oyun.cizmek()

    pygame.display.update()  # Ekranı güncelle
    saat.tick(FPS)  # FPS'i sınırla
    
    
    
    

pygame.quit()  # Pygame'i kapat
```
